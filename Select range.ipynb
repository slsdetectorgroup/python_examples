# Select range from two arrays

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.arange(0, 2000, 20)
y = x**2
print(f'{x=}\n{y=}')
```

```python
#using boolean mask
mask = tuple([y<200e3]) #need tuple of bool to mask array
fig, ax = plt.subplots()
ax.plot(x[mask], y[mask])
```

```python
#using array indicies
idx = np.where(y>200e3)[0][0] # np.where returns a tuple of arrays
fig, ax = plt.subplots()
ax.plot(x[0:idx], y[0:idx])
idx
```
